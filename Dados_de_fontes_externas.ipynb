```python
# importação do conjuntos de dados iris
from sklearn.datasets import load_iris

# importação de bibliotecas para manipulação dos dados
import pandas as pd
```

```python
iris = load_iris()
iris.data
```

```python
# Descrição do conjunto de dados
print(iris.DESCR)
```

```python
# Metadados do dataset
iris = pd.Series(iris)
iris
```

```python
feature_names = iris.feature_names
feature_names
```

```python
# transfomando modelo dos dados para um Dataframe
iris_df = pd.DataFrame(iris.data, 
                  columns = feature_names)
iris_df
```
